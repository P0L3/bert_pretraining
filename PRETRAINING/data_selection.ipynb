A notebook to select and export a specific sample of the PDF_TXT data.

## Data and Libraries

```python
import pandas as pd
```

```python
REF_FILE = "DATASET/ED4RE_2503/ED4RE_2603.pickle"

df = pd.read_pickle(REF_FILE)

# Add for ILOC
df["backtrace_index"] = range(len(df))
```

```python
df.info()
```

```python
df.Journal.unique()
```

```python
SELECTED_IDS = df[df.Journal.isin(["Climate", "Ecologies", "Environments"])]["backtrace_index"].to_list()
```

```python
len(SELECTED_IDS)
```

```python
import os
DIR = "/home/p0l3/RAD/REL_DIS/REL_DIS/TOKENIZED/PAPERS_SENTENCES"

file_list = os.listdir(DIR)

sample_list = [f for f in file_list if int(f.split("_")[0]) in SELECTED_IDS]
```

```python
len(sample_list)
```

```python
import shutil


# Define the source directory where the files currently are
source_dir = DIR # '/home/p0l3/RAD/CWED4ETA/CWED4ETA/CWED4ETA/RESULTS/GLINER_TESTS_FULL'

# Define the destination directory where you want to copy the files
destination_dir = "/home/p0l3/RAD/BERT_PRETRAINING/PRETRAINING/DATASET/RIBARIC" # '/home/p0l3/RAD/CWED4ETA/CWED4ETA/CWED4ETA/RESULTS/GLINER_TESTS_SAMPLE'

# Create the destination directory if it doesn't exist
os.makedirs(destination_dir, exist_ok=True)

# Loop through the list and copy each file
for file_name in sample_list:
    # Create the full path to the source file
    source_file = os.path.join(source_dir, file_name)
    
    # Create the full path to the destination file
    destination_file = os.path.join(destination_dir, file_name)
    
    # Copy the file
    shutil.copy(source_file, destination_file)
    print(f"Copied {file_name} to {destination_dir}")
```
